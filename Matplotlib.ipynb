```python
import matplotlib.pyplot as plt
import numpy as np

# Define X from 0 to 20
x = np.linspace(0, 20, 100)

# Define the three functions
y1 = x           # y = x
y2 = x**2        # y = x^2
y3 = x**3        # y = x^3

# Plot the lines
plt.plot(x, y1, label='y = x', color='blue')
plt.plot(x, y2, label='y = x²', color='green')
plt.plot(x, y3, label='y = x³', color='red')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graphs of y = x, y = x², y = x³')
plt.legend()


```
